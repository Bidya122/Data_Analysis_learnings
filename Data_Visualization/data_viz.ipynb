box and violin plot

```python
# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

# For handling data (like gene expression tables)
import pandas as pd
import numpy as np

# Example dataset: gene expression of 100 samples across 5 genes
np.random.seed(42)  # reproducibility, u can take any number instead of 42.
df = pd.DataFrame({
    "Condition": np.random.choice(["Control", "Treatment"], size=100),
    "Gene1": np.random.normal(10, 2, 100),   # mean=10, sd=2
    "Gene2": np.random.normal(8, 1.5, 100),
    "Gene3": np.random.normal(12, 3, 100),
    "Gene4": np.random.normal(6, 2, 100),
    "Gene5": np.random.normal(15, 2.5, 100),
})

df.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="Condition", y="Gene1", data=df)
sns.violinplot(x="Condition", y="Gene1", data=df, color="lightgray", alpha=0.3)
plt.title("Gene1 expression across Control vs Treatment")
plt.show()
```

```python
genes = ["Gene1", "Gene2", "Gene3", "Gene4", "Gene5"] ##took value for all genes then assigned for each of the genes

for gene in genes:
    plt.figure(figsize=(8,6))
    sns.boxplot(x="Condition", y=gene, data=df)
    sns.violinplot(x="Condition", y=gene, data=df, color="lightgray", alpha=0.3)
    plt.title(f"{gene} expression across Control vs Treatment")
    plt.ylabel("Expression Level")
    plt.show()
```

```python
df_melted = df.melt(id_vars="Condition", value_vars=["Gene1","Gene2","Gene3","Gene4","Gene5"],
                    var_name="Gene", value_name="Expression")

plt.figure(figsize=(10,6))
sns.boxplot(x="Gene", y="Expression", hue="Condition", data=df_melted,
            palette={"Control": "lightgreen", "Treatment": "darkgreen"})
sns.violinplot(x="Gene", y="Expression", hue="Condition", data=df_melted,
               palette={"Control": "lightgray", "Treatment": "silver"},
               alpha=0.3, dodge=True)

plt.title("Expression of all genes across Control vs Treatment")
plt.ylabel("Expression Level")
plt.xlabel("Genes")
plt.legend(title="Condition")
plt.show()
```

scatter plot

```python
# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

# For handling data (like gene expression tables)
import pandas as pd
import numpy as np

# Example dataset: gene expression of 100 samples across 5 genes
np.random.seed(42)  # reproducibility, u can take any number instead of 42.
df = pd.DataFrame({
    "Condition": np.random.choice(["Control", "Treatment"], size=100),
    "Gene1": np.random.normal(10, 2, 100),   # mean=10, sd=2
    "Gene2": np.random.normal(8, 1.5, 100),
    "Gene3": np.random.normal(12, 3, 100),
    "Gene4": np.random.normal(6, 2, 100),
    "Gene5": np.random.normal(15, 2.5, 100),
})

plt.figure(figsize=(8,6))
sns.scatterplot(x="Gene1", y="Gene2", hue="Condition", data=df, s=80, alpha=0.7)
plt.title("Scatter Plot: Gene1 vs Gene2")
plt.xlabel("Gene1 Expression")
plt.ylabel("Gene2 Expression")
plt.legend(title="Condition")
plt.show()
```

```python
sns.pairplot(df, vars=["Gene1","Gene2","Gene3","Gene4","Gene5"], hue="Condition")
plt.suptitle("Pairwise Scatter Plots of Gene Expression", y=1.02)
plt.show()
```

```python
long = df.melt(id_vars="Condition",
               value_vars=["Gene1","Gene2","Gene3","Gene4","Gene5"],
               var_name="Gene", value_name="Expression")

plt.figure(figsize=(10,6))
sns.scatterplot(x="Gene", y="Expression", hue="Condition", data=long, s=80, alpha=0.7)
plt.title("Expression Levels of All Genes by Condition")
plt.xlabel("Gene")
plt.ylabel("Expression Level")
plt.show()
```

3D Plot

```python
# Import plotting libraries
import matplotlib.pyplot as plt
import seaborn as sns

# For handling data (like gene expression tables)
import pandas as pd
import numpy as np

# Example dataset: gene expression of 100 samples across 5 genes
np.random.seed(42)  # reproducibility, u can take any number instead of 42.
df = pd.DataFrame({
    "Condition": np.random.choice(["Control", "Treatment"], size=100),
    "Gene1": np.random.normal(10, 2, 100),   # mean=10, sd=2
    "Gene2": np.random.normal(8, 1.5, 100),
    "Gene3": np.random.normal(12, 3, 100),
    "Gene4": np.random.normal(6, 2, 100),
    "Gene5": np.random.normal(15, 2.5, 100),
})

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection="3d")

# 3D scatter plot
ax.scatter(df["Gene1"], df["Gene2"], df["Gene3"],
           c=df["Condition"].map({"Control":"blue", "Treatment":"red"}),
           s=60, alpha=0.7)

# Axis labels
ax.set_xlabel("Gene1 Expression")
ax.set_ylabel("Gene2 Expression")
ax.set_zlabel("Gene3 Expression")
plt.title("3D Scatter Plot of Gene1, Gene2, and Gene3")

plt.show()
```

Histogram

```python
plt.figure(figsize=(8,6))
sns.histplot(df['Gene1'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Gene1 Expression')
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.show()
```

```python
df_melted = df.melt(id_vars='Condition',
                    value_vars=['Gene1','Gene2','Gene3','Gene4','Gene5'],
                    var_name='Gene',
                    value_name='Expression')

plt.figure(figsize=(10,6))
sns.histplot(data=df_melted, x='Expression', hue='Gene', element='step', bins=15, kde=True)
plt.title('Distribution of Expression for All 5 Genes')
plt.xlabel('Expression Level')
plt.ylabel('Frequency')
plt.show()
```

```python
genes = ['Gene1','Gene2','Gene3','Gene4','Gene5']

fig, axes = plt.subplots(1, 5, figsize=(20,4), sharey=True)  # 1 row, 5 columns, same y-axis

for i, gene in enumerate(genes):
    sns.histplot(df[gene], bins=15, kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(gene)
    axes[i].set_xlabel('Expression')
    if i == 0:
        axes[i].set_ylabel('Frequency')
    else:
        axes[i].set_ylabel('')

plt.suptitle('Histograms of All 5 Genes', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(x='Condition', data=df, palette='pastel')
plt.title('Number of Samples per Condition')
plt.xlabel('Condition')
plt.ylabel('Count')
plt.show()
```

Heat Maps

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Take only the gene expression columns
genes = ["Gene1", "Gene2", "Gene3", "Gene4", "Gene5"]

plt.figure(figsize=(8,6))
sns.heatmap(df[genes], cmap="viridis", cbar=True)
plt.title("Heatmap of Gene Expression (all samples)")
plt.xlabel("Genes")
plt.ylabel("Samples")
plt.show()
```

pie charts

```python
# Average expression of all 5 genes
gene_means = df[["Gene1","Gene2","Gene3","Gene4","Gene5"]].mean()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(gene_means,
        labels=gene_means.index,
        autopct="%1.1f%%",
        startangle=90,
        colors=sns.color_palette("pastel"))
plt.title("Average Expression Contribution of Each Gene")
plt.show()
```

```python
grouped = df.groupby("Condition")[["Gene1","Gene2","Gene3","Gene4","Gene5"]].mean()
print(grouped)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,6))

# Pie for Control
axes[0].pie(grouped.loc["Control"],
            labels=grouped.columns,
            autopct="%1.1f%%",
            startangle=90,
            colors=sns.color_palette("pastel"))
axes[0].set_title("Gene Expression Distribution - Control")

# Pie for Treatment
axes[1].pie(grouped.loc["Treatment"],
            labels=grouped.columns,
            autopct="%1.1f%%",
            startangle=90,
            colors=sns.color_palette("pastel"))
axes[1].set_title("Gene Expression Distribution - Treatment")

plt.show()
```

```python
# Reshape data
df_melt = df.melt(id_vars="Condition",
                  value_vars=["Gene1","Gene2","Gene3","Gene4","Gene5"],
                  var_name="Gene", value_name="Expression")

plt.figure(figsize=(8,6))
sns.barplot(x="Gene", y="Expression", hue="Condition", data=df_melt,
            ci="sd", palette="pastel")
plt.title("Average Gene Expression with Error Bars (SD)")
plt.xlabel("Genes")
plt.ylabel("Expression")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x="Condition", y="Gene1", data=df,
            ci="sd",  # ci="sd" means show standard deviation as error bars
            palette="pastel")
plt.title("Gene1 Expression with Error Bars")
plt.xlabel("Condition")
plt.ylabel("Expression")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.pointplot(x="Gene", y="Expression", hue="Condition",
              data=df_melt,
              ci="sd",          # error bars = standard deviation
              dodge=True,       # separates Control vs Treatment
              markers="o",
              linestyles="-",
              palette="pastel")

plt.title("Gene Expression Line Plot with Error Bars")
plt.xlabel("Gene")
plt.ylabel("Expression")
plt.show()
```
